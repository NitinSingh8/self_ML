## Line Graph

```python
# graph of x = y^2

x_axis = np.arange(0,20)
y_axis = x_axis**2


# plt.plot(x_axis,y_axis,'g.')  # g for color(green) , + is type of line you want
plt.plot(x_axis,y_axis,color='green', marker='+', linestyle='dashed',linewidth=2, markersize=8)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph of x = y^2')
plt.show()
```

```python
# graph of sin , cos , tan

x = np.arange(0,10,0.2)

sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)

# graph of sin function
plt.plot(x,sin_y)
plt.title("Sin graph")
plt.show()

# graph of cos function
plt.plot(x,cos_y)
plt.title("Cos graph")
plt.show()

# graph of tan function
plt.plot(x,tan_y)
plt.title("Tan graph")
plt.show()
```

```python
# combine thre graph in one figure (sin,cos,tan)

x = np.arange(0,10,0.5)

sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)

plt.plot(x,sin_y,label='sin')    # graph of sin function
plt.plot(x,cos_y,label='cos')   # graph of cos function
plt.plot(x,tan_y,label='tan')   # graph of tan function
plt.title("Graph of sin , cos ,tan")
plt.legend()
plt.show()
```

# Subplot

```python
# let's make some graph and plot in a figure

x = np.arange(0,2,0.1)
a = x
b = x**2
c = x**3
d = x**4


plt.figure(figsize=(10,10)) 
 
# now let's make graph of (x and a), (x and b), (x and c), (x and d)

plt.subplot(221)  # subplot(221) -> means 22= 2*2 matrix and 1 means first position  (make 2*2 matrix and my graph should be at first position out of 4 position)
plt.plot(x,a)
plt.title("x=a")

plt.subplot(222)
plt.plot(x,b)
plt.title("x=a^2")

plt.subplot(223)
plt.plot(x,c)
plt.title("x=a^3")

plt.subplot(224)
plt.plot(x,d)
plt.title("x=a^4")

plt.show()
```

# Scatter Plot

```python
# parabola with scatter plot
# y = x^2
x = np.arange(-10,10,0.5)
y = x**2
plt.scatter(x,y)
plt.title("Scatter plot of parabola")
plt.show()
```

# Bar Graph/Chart

```python
# how many people likes which fruit
fruits = ['Apple','Orange','Banana','Kiwifruit','Blueberry','Grapes']
people = [55,25,40,20,45,10]


plt.bar(fruits,people)
plt.xlabel("Fruits")
plt.ylabel("No. of People likes")
plt.show()
```

```python
# how many people likes which fruit
fruits = ['Apple','Orange','Banana','Kiwifruit','Blueberry','Grapes']
people = [55,25,40,20,45,10]


plt.barh(fruits,people)
plt.xlabel("Fruits")
plt.ylabel("No. of People likes")
plt.show()
```

```python
# bar graph with multiple data
from matplotlib import style
style.use('ggplot') # for just changing the background


index = [1,2,3,4] # for x axis

a = [10,20,30,40] # for y axis
b = [20,10,35,38] # for y axis

plt.bar(index,a,color = "green") # you can also give width and increase or decrease
plt.bar(index,b,color = "red")
plt.show()
```

# Pie Chart

```python
# how many people likes which fruit
fruits = ['Apple','Orange','Banana','Kiwifruit','Blueberry','Grapes']
people = [55,25,40,20,45,10]


plt.pie(people,labels = fruits,autopct='%1.1f%%')
plt.show()
```

```python
marks = np.array([45,24,63,42,55])
names=  np.array(["Rohan","Shivam","Rajesh","Amit","Aryan"])
colors = ["Red","Blue","Green","Red","Yellow"]
explodes = [0,0,0.2,0,0]
plt.pie(marks,labels = names,explode = explodes,colors = colors,shadow = True,autopct='%1.1f%%')
plt.title("All in one pie chart")
plt.show()
```
